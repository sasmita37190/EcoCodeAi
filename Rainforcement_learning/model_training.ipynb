```python
import gymnasium as gym
import numpy as np
import pandas as pd

class CodeEfficiencyEnv(gym.Env):
    def __init__(self, df):
        super(CodeEfficiencyEnv, self).__init__()
        self.df = df
        self.index = 0
        
        # Define Action Space (0: Keep Code, 1: Suggest Optimization)
        self.action_space = gym.spaces.Discrete(2)

        # Define Observation Space with realistic upper bounds
        self.observation_space = gym.spaces.Box(
            low=0, 
            high=np.array([100000, 100000, 10, 1000, 100, 50]),  # Adjusted realistic values
            shape=(6,), 
            dtype=np.float32
        )

    def get_state(self):
        """Extracts the current state from the dataset."""
        row = self.df.iloc[self.index]
        return np.array([
            row['CPU_Time(ms)'], 
            row['Memory(KB)'], 
            row['Energy(kWh)'],
            row['Carbon_Emission(gCO₂)'], 
            row['Loops'], 
            row['Control_Flows']
        ], dtype=np.float32)

    def step(self, action):
        """Executes an action and returns the next state, reward, and termination flag."""
        reward = 1 if action == self.df.iloc[self.index]['Efficiency_Label'] else -1
        
        self.index += 1
        terminated = self.index >= len(self.df)  # Done when all data points are processed
        truncated = False  # Modify this if you want an early stopping condition

        if not terminated:
            state = self.get_state()
        else:
            state = np.zeros(self.observation_space.shape, dtype=np.float32)  # Return a zero-state instead of None
        
        return state, reward, terminated, truncated, {}  # Returning 5 values as per Gym API

    def reset(self, seed=None):
        """Resets the environment and returns the initial state."""
        super().reset(seed=seed)  # Properly handle seeding
        self.index = 0
        return self.get_state(), {}  # Return initial state and empty info dict
```

```python
# Initialize environment
env = CodeEfficiencyEnv(df)

# Reset environment and get initial state
state, _ = env.reset()
done = False

while not done:
    action = np.random.choice([0, 1])  # Randomly selecting an action (0: Keep, 1: Optimize)
    state, reward, terminated, truncated, _ = env.step(action)  # Expect 5 values
    done = terminated or truncated  # Loop stops if either is True

    print(f"Action: {action}, Reward: {reward}")


```

```python
import gymnasium as gym  # Use gymnasium instead of gym
import numpy as np
import pandas as pd
from stable_baselines3 import DQN  # Deep Q-Learning Algorithm
from stable_baselines3.common.env_checker import check_env
from sklearn.model_selection import train_test_split
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print(f"Training Samples: {len(train_df)}")
print(f"Testing Samples: {len(test_df)}")
```

```python
env = CodeEfficiencyEnv(train_df)
check_env(env)  # Optional: To verify the environment is built correctly
```

```python
model = DQN("MlpPolicy", env, verbose=1, learning_rate=0.001, gamma=0.99)
model.learn(total_timesteps=100000)
model.save("eco_code_ai_model")
print("Training Completed ✅")
```

```python
test_env = CodeEfficiencyEnv(test_df)
state, _ = test_env.reset()  # Unpack the tuple correctly
done = False
total_reward = 0

while not done:
    action, _ = model.predict(state)  # Now `state` is correctly formatted
    state, reward, done, _, _ = test_env.step(action)  # Unpack correctly
    total_reward += reward

print(f"Total Reward: {total_reward}")
```

```python
accuracy = (total_reward + len(test_df)) / (2 * len(test_df))
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
import torch.nn as nn

class CodeClassifier(nn.Module):
    def __init__(self, input_size):
        super(CodeClassifier, self).__init__()
        self.fc1 = nn.Linear(input_size, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 2)  # ⚠️ Ensure 2 output neurons

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return torch.softmax(self.fc3(x), dim=1)  # ⚠️ Softmax for classification

```

```python
import torch
from stable_baselines3 import DQN
import re

def extract_features_from_code(code_string):
    """Full feature extraction with 6 features as the model expects"""
    # Basic metrics
    length = len(code_string)
    num_loops = len(re.findall(r'\b(for|while)\b', code_string))
    num_conditions = len(re.findall(r'\b(if|elif|else)\b', code_string))
    num_functions = code_string.count("def ")
    num_variables = len(re.findall(r'\b\w+\s*=', code_string))
    
    # Recursion detection
    function_match = re.search(r'def\s+(\w+)\(', code_string)
    num_recursive = 0
    if function_match:
        func_name = function_match.group(1)
        num_recursive = max(len(re.findall(fr'\b{func_name}\s*\(', code_string)) - 1, 0)

    # Complexity calculation (matches original training features)
    complexity_score = (
        num_loops * 2.0 +
        num_recursive * 7.0 +
        num_variables * 0.5 +
        num_conditions * 1.5
    )

    # 6 features with proper normalization
    return [
        length / 10000,          # Feature 1: Normalized length
        num_loops / 15,          # Feature 2: Loops
        num_conditions / 10,     # Feature 3: Conditionals
        num_functions / 5,       # Feature 4: Functions
        num_variables / 20,      # Feature 5: Variables
        complexity_score / 100   # Feature 6: Complexity
    ]

def evaluate_code_efficiency(code_string):
    """Fixed observation shape evaluation"""
    try:
        # Extract features and ensure correct shape
        features = extract_features_from_code(code_string)
        obs = torch.tensor(features, dtype=torch.float32).view(1, -1)  # Shape [1, 6]
        
        # Get prediction
        action, _ = model.predict(obs.numpy(), deterministic=True)
        
        return "Efficient Code ✅" if action == 1 else "Inefficient Code ❌"
    except Exception as e:
        return f"Error: {str(e)}"

if __name__ == "__main__":
    try:
        # Load model (ensure correct path)
        model = DQN.load(r"C:\Users\Deba Prakash\Desktop\Model Traning\eco_code_ai_model.zip")
        
        # Test with Fibonacci code
        test_code = "def hello(): print(Hello World)"
        print(evaluate_code_efficiency(test_code))
    except Exception as e:
        print(f"Runtime Error: {e}")
```
